```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
raw_df = pd.read_csv('Trading_Project_Data.csv')
raw_df
```

```python
"""Data Preprocessing"""
df = pd.read_csv('Trading_Project_Data.csv', header=[0,1], skiprows=[2])
df.columns = [f"{col[0]}_{col[1]}" for col in df.columns]
df['Price_Ticker']=pd.to_datetime(df['Price_Ticker'], dayfirst=True)
df.set_index('Price_Ticker', inplace=True)
final = df.dropna()
final
```

```python
final.to_csv('Trading_Project_Data_Cleaned.csv',index=True)
```
